```python
word=input("Enter a word lesser than 10 letters: ")
guess=""
hangman=[]
while len(word)>10:
  print("please enter another word")
  word=input("Enter a word lesser than 10 letters: ")
word=word.lower()
print(f"There are {len(word)} letters: ")
for i in range(len(word)):
  hangman.append("")
for i in range (10):
  letter=input("Enter a lowercase letter: ")
  found=False
  for j in range(len(word)):
    if letter==word[j]:
      print(f"Your letter is in the {j} position")
      hangman[j]=letter
      found=True
  if found==False:
    print("The letter is not in the word")
  else:
    print(hangman)
    guess=input("Enter your guess: ")
    guess=guess.lower()
    if guess==word:
      print("You have guessed it correctly")
      break
    else:
      print("incorrect enter a new letter")
if guess!=word:
  print("You have entered all ten letters")
  answer=input("What is the answer: ")
  answer=answer.lower()
  if answer==word:
    print("That is the correct word,You win!")
  else:
    print("That is not the correct word,player one wins!")
```
